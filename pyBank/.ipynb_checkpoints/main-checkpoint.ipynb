```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path('data.csv')

# Open the csv file as an object
with open(csvpath, 'r') as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)
    

    #Intializing variables
    #To calculate number of months
    month_count = 0
    
    #To calculate sum of all profit and loss amounts
    total_profitloss = 0
    
    #To calculate average change in profir and loss
    total_change_in_profit_loss = 0
    
    #To calculate greatest increase and greatest decrease in profit
    greatest_increase_in_profits = 0
    greatest_decrease_in_profits = 0
    
    

    #Read each row after the header
    for row in csvreader:        
        
        #Set the 'date', and 'profit/losses' variables
        #Convert profit/losses to integer for numerical calculations
        month = row[0]
        profit_or_loss = int(row[1])
        
        #calculate the total change in profit or loss by adding
        #the changes for each month 
        
        if month_count != 0:
            change_in_profit_loss = (profit_or_loss - previous_month_profit_or_loss)
            total_change_in_profit_loss += change_in_profit_loss
            
         #If the change in profit or loss for a period is greater than the
         #pre-calculated greatest increase, assign the same to greatest increase
        
            if change_in_profit_loss > greatest_increase_in_profits:
                greatest_increase_in_profits = change_in_profit_loss
                
         #If the change in profit or loss for a period is less than the
         #pre-calculated greatest decrease, assign the same to greatest decrease      
        
            elif change_in_profit_loss < greatest_decrease_in_profits:
                greatest_decrease_in_profits = change_in_profit_loss
        
        #Count this iteration as another month    
        month_count += 1
        
        #Assign present month's profit or loss to Previous month for the next iteration        
        previous_month_profit_or_loss = profit_or_loss
        
        #Add present month's profit or loss to the cumulated value        
        total_profitloss += profit_or_loss
        
        
    
    # Print the header for the output
    print("Financial Analysis")
    print("-------------------")
    
    print(f"Total number of Months = {month_count}")
    print(f"Total amount of Profit/Losses = ${total_profitloss}")
    
    #Compute average change by dividing total change by number of change values
    average_change_in_profit_and_loss = total_change_in_profit_loss / (month_count - 1)
    
    print(f"Average change in Profit/Loss = ${average_change_in_profit_and_loss:.2f}")
    
    print(f"Greatest increase in profits = ${greatest_increase_in_profits}")
    
    print(f"Greatest debrease in profits = ${greatest_decrease_in_profits}")
    
    
   

```
